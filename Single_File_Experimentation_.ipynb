```python
# Membuat Automatisasi pembuka ekstensi File
# EEG_DATA, MEG_DATA, ELECTRODES DATA
directory = "/content/"
for file in os.listdir(directory):
    file_path = os.path.join(directory, file)
    if os.path.isfile(file_path):
        file_extension = os.path.splitext(file_path)[1]
        print(file, "ends in", file_extension)
        for x in list_filenya:
                  if file_extension.lower() == ".vhdr":
                     raw_eeg = mne.io.read_raw_brainvision(x, preload=True)
                  elif file_extension.lower() == ".cnt":
                     raw_eeg = mne.io.read_raw_cnt(x, preload=True)
                  elif file_extension.lower() == ".edf":
                     raw_eeg = mne.io.read_raw_edf(x, preload=True)
                  elif file_extension.lower() == ".bdf":
                     raw_eeg = mne.io.read_raw_bdf(x, preload=True)
                  elif file_extension.lower() == ".gdf":
                     raw_eeg = mne.io.read_raw_edf(x, preload=True)
                  elif file_extension.lower() == ".egi":
                     raw_eeg = mne.io.read_raw_egi(x, preload=True)
                  elif file_extension.lower() == ".mff":
                     raw_eeg = mne.io.read_raw_egi(x, preload=True)
                  elif file_extension.lower() == ".set":
                     raw_eeg = mne.io.read_raw_eeglab(x, preload=True)
                  elif file_extension.lower() == ".nxe":
                     raw_eeg = mne.io.read_raw_eximia(x, preload=True)
                  elif file_extension.lower() == ".fif":
                     raw_eeg = mne.io.read_raw_fif(x, preload=True)
                  elif file_extension.lower() == ".bti":
                     raw_eeg = mne.io.read_raw_bti(x, preload=True)
                  elif file_extension.lower() == ".ctf":
                     raw_eeg = mne.io.read_raw_ctf(x, preload=True)
                  elif file_extension.lower() == ".kit":
                     raw_eeg = mne.io.read_raw_kit(x, preload=True)
                  else:
                     print('File Kurang Sinkron')
```

```python
# ICA untuk menghilangkan Aftefak (Biasanya melalui EOG)
# Karena ingin menghilangkan channel jadi kita dekomposisi jadi 20 saja
ica = mne.preprocessing.ICA(n_components=20, random_state=0)
```

```python
# Mengaplikasikan 
```

```python
# Setting Filter Bandpass-Filter (Low + High)
low_filter = 0.5
high_filter = 50
```

```python
# Mengambil Sampling Frequency terlebih dahulu
sfreq = raw_eeg.info.get('sfreq')
```

```python
# Melakukan import library ekstraksi fitur
from mne_features.feature_extraction import extract_features
import mne_features
```

```python
# Melakukan fitlering juga dengan bandpass filter, dimana menghilangkan 0.5-50 Hz
raw_2 = raw_eeg.filter(low_filter, high_filter)
```

```python
# Transpose data
eeg_baru_filter = raw_2.get_data().T
#Melakukan Set Channel Montage
montage = mne.channels.make_standard_montage('standard_1020')
raw_2.set_montage(montage, on_missing='ignore')

# Membuat nama Channel
channels = raw_2.info.ch_names

# Ubah data dari vdhr ke dataframe
import pandas as pd
df_baru_filter = pd.DataFrame(eeg_baru_filter, columns=channels)
```

```python
# Melakukan ICA FIT
ica.fit(raw_2.copy().filter(8, 35))
```

```python
# Jangan lupa melakukan ICA pada komponen channel artefak untuk menghilangkannya 
# misalnya mata menggunakan EOG untuk menghilangkan artefak pergerakan mata
# EMG untuk acuan menghilangkan artefak berbasis gerakan otot
# EKG/ECG untuk acuan menghilangkan artefak berbasis detak jantung
import re       
EOG = list(df_baru_filter.columns[df_baru_filter.columns.str.contains('EOG')])
ECG = list(df_baru_filter.columns[df_baru_filter.columns.str.contains('ECG')])
EMG = list(df_baru_filter.columns[df_baru_filter.columns.str.contains('EMG')])
Artefak_all=[EOG, ECG, EMG]

# Karena MNE hanya support remove atricact berbasis EOG dan ECG saja, maka hanya ini yang bisa dilakukan
# Cek, jika list mengandung komponen string yang dimaksud, maka masukkan sebagai indikator
result_eog = any(len(elem_eog) > 0 for elem_eog in EOG)
if result_eog:
   bad_id_eog, scores_artef_eog = ica.find_bads_eog(raw_2, EOG, threshold=2)
   print(bad_id_eog)   
else:
   print('Tidak ditemukan EOG Channel')
 
result_ecg = any(len(elem_ecg) > 0 for elem_ecg in ECG)
if result_ecg:
   bad_id_ecg, scores_artef_ecg = ica.find_bads_ecg(raw_2, ECG, method='correlation', threshold=2)
   print(bad_id_ecg)
else :
  print('Tidak ditemukan ECG Channel')
```

```python
# Melihat semua ID artefak
import itertools

# Melakukan kompilasi jika ada dua artefak (EOG dan ECG)
if result_ecg and result_eog:
   all_artif = list(itertools.chain(bad_id_eog, bad_id_ecg))
   sorted_all_artif = sorted(all_artif)
   duplic_id_artif = []
   for all_elem in sorted_all_artif:
         if sorted_all_artif.count(all_elem)>1:
            if all_elem not in duplic_id_artif:
               duplic_id_artif.append(all_elem)
   print(duplic_id_artif)
   clean_artif = [x for x in all_artif if x not in [duplic_id_artif]]
   ica.exclude= clean_artif
elif result_ecg and not result_eog:
    ica.exclude= bad_id_ecg 
elif result_eog and not result_ecg:
    ica.exclude= bad_id_eog
else : 
    print('gak ada')
```

```python
# Mengaplikasikan ICA pada dataset
eeg_with_ica = ica.apply(raw_2.copy(), exclude=ica.exclude)
```

```python
eeg2 = eeg_with_ica.get_data().T
df_bersih = pd.DataFrame(eeg2, columns=channels)
```

```python
# Memulai Feature Extraction
```

```python
# Mengambil Power Spectral Density
# Normalisasikan PSD
psds, freqs = mne.time_frequency.psd_welch(eeg_with_ica)
psds_norm = psds.sum(axis=1)
print(psds_norm)
```

```python
# Mengambil Mean, Median, dan Variance masing-masing channel
mean_each_chan = mne_features.univariate.compute_mean(eeg_with_ica.get_data())
var_each_chan = mne_features.univariate.compute_variance(eeg_with_ica.get_data())
skew_each_chan = mne_features.univariate.compute_skewness(eeg_with_ica.get_data())
kurt_each_chan = mne_features.univariate.compute_kurtosis(eeg_with_ica.get_data())
H_frac_dim_chan = mne_features.univariate.compute_higuchi_fd(eeg_with_ica.get_data(), kmax=10)
K_frac_dim_chan = mne_features.univariate.compute_katz_fd(eeg_with_ica.get_data())
Zero_Cross_chan = mne_features.univariate.compute_zero_crossings(eeg_with_ica.get_data(), threshold=2.220446049250313e-16)
Hj_mobility_chan = mne_features.univariate.compute_hjorth_mobility(eeg_with_ica.get_data())
Hj_complexity_chan = mne_features.univariate.compute_hjorth_complexity(eeg_with_ica.get_data())
decor_chan = mne_features.univariate.compute_decorr_time(sfreq, eeg_with_ica.get_data())
```

```python
# Feature Extraction
# 1. Mean, Median, Variance keseluruhan dataframe
# 1. Mean, Median, dan Variance masing-masing channel 
# 2. Kurtosis, Skewness masing-masing channel
# 3. Zero Crossing Rate masing-masing channel
# 4. Shannon Entropy masing-masing channel
# 5. Power Sprectral Density
# 6. Renyi Entropy
# 7. Distribution of Brain Waves (Delta, Theta, Alpha, Gamma) [Mengambil mean, median, dan juga variance masing-masing gelombang]
```

```python
shan_entropy_chan = mne_features.univariate.compute_spect_entropy(sfreq, eeg_with_ica.get_data(), psd_method='welch', psd_params=None)
line_length_chan = mne_features.univariate.compute_line_length(eeg_with_ica.get_data())
```
